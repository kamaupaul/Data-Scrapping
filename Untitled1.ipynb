```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://cytonn.com/blog/article/cytonn-top-employer'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
content = soup.find_all('tbody')
print(content)
```

```python
titles = soup.find_all('tr') 
print(titles)
```

```python
for tr in titles:
    title = [title.text.strip() for title in tr]
    df = [title[1],title[3],title[5],title[7]]
    print(df)
```

```python
import pandas as pd
data = pd.DataFrame(df, columns=['Rank1', 'Company1', 'Rank2', 'Company2'])

data
```
